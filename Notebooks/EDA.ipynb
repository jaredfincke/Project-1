# Exploratory Data Analysis

```python
import pandas as pd
Tdf = pd.read_csv("../data/Teams_Raw.csv")
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Looking at a summary of the four main variables

```python
print(Tdf['W'].describe())
print(Tdf['E'].describe())
print(Tdf['ER'].describe())
print(Tdf['SB'].describe())
```

### Visualizing the distribution of the four main variables

```python
TNdf = Tdf[['W','E','R','ER','SB']]
TNdf.hist(figsize=(16,20), bins=50, xlabelsize=8, ylabelsize=8)
```

```python
### Plotting the independent variables to the dependent variable (Wins)
### Performing a sensibility test by eyeing the direction of the perceived slope.
```

```python
TNdf.plot(kind = 'scatter', x = 'E', y = 'W')
TNdf.plot(kind = 'scatter', x = 'R', y = 'W')
TNdf.plot(kind = 'scatter', x = 'ER', y = 'W')
TNdf.plot(kind = 'scatter', x = 'SB', y = 'W')
plt.show()
```

### Comparing the means of my independent variables between World Series Winners and Non World Series Winners
